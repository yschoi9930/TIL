```python
html = driver.page_source
```

```python
soup = BeautifulSoup(html)
```

```python
name = soup.find_all('div','name CareName')
```

```python
temp = soup.find_all('div','capacity')
temp
```

```python
names = []
for i in range(0,138) :
    temp = soup.find_all('div', 'name CareName')
    name = temp[i].a.text
    names.append(name)
print(names)
```

```python
addresses = []
for i in range(0,138) :
    temp = soup.find_all('div', 'address CareAddress')
    name = temp[i].text
    addresses.append(name)
print(addresses)
```

```python
capacity = []
for i in range(0,138) :
    temp = soup.find_all('div', 'capacity')
    name = temp[i].text
    capacity.append(name)
print(capacity)
```

```python
data = pd.DataFrame({'이름' : names,'주소' : addresses,'정원':capacity})
```
